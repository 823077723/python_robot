```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 输入点的坐标
points = np.array([
    [1, 1, 3],
    [4, 5, 6],
    [7, 8, 9]
])

# 计算平面方程
P1, P2, P3 = points[0], points[1], points[2]
v1 = P2 - P1
v2 = P3 - P1
normal = np.cross(v1, v2)  # 计算法向量
A, B, C = normal

# 求D的值
D = np.dot(normal, P1)

# 平面方程
print(f"平面方程: {A}x + {B}y + {C}z = {D}")
print(f"法向量: {normal}")


# 创建三维图形
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 绘制点
ax.scatter(points[:, 0], points[:, 1], points[:, 2], color='r', s=50)

# 绘制平面
x = np.linspace(min(points[:, 0]), max(points[:, 0]), 10)
y = np.linspace(min(points[:, 1]), max(points[:, 1]), 10)
X, Y = np.meshgrid(x, y)
Z = (D - A * X - B * Y) / C  # 根据平面方程计算Z

ax.plot_surface(X, Y, Z, alpha=0.9, rstride=100, cstride=100, color='b')

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 输入点的坐标
points = np.array([
    [1, 0, 0],
    [0, 5, 0],
    [0, 0, 9]
])

# 计算平面方程
P1, P2, P3 = points[0], points[1], points[2]
v1 = P2 - P1
v2 = P3 - P1
normal = np.cross(v1, v2)  # 计算法向量
A, B, C = normal

# 求D的值
D = np.dot(normal, P1)

# 平面方程
print(f"平面方程: {A}x + {B}y + {C}z = {D}")

# 创建一个含有 2x2 子图的图形
fig = plt.figure(figsize=(10, 10))

# 设置多个视角
views = [(30, 45), (60, 90), (90, 135), (120, 180)]  # 4个不同视角

for i, (elev, azim) in enumerate(views):
    # 创建子图
    ax = fig.add_subplot(2, 2, i+1, projection='3d')  # 2行2列的子图
    ax.set_title(f'视角 {i+1}: elev={elev}, azim={azim}')
    
    # 绘制点
    ax.scatter(points[:, 0], points[:, 1], points[:, 2], color='r', s=50)
    
    # 绘制平面
    x = np.linspace(min(points[:, 0]), max(points[:, 0]), 10)
    y = np.linspace(min(points[:, 1]), max(points[:, 1]), 10)
    X, Y = np.meshgrid(x, y)
    Z = (D - A * X - B * Y) / C  # 根据平面方程计算Z
    ax.plot_surface(X, Y, Z, alpha=0.1, rstride=100, cstride=100, color='b')

    # 设置视角
    ax.view_init(elev=elev, azim=azim)

# 显示所有视角的图形
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from itertools import combinations

# 输入点的坐标 (n = 4)
points = np.array([
    [1, 0, 1],
    [0, 5, 1],
    [0, 0, 1],
    [2, 2, 0],
    [6, 5, 1]
])

# 创建三维图形
fig = plt.figure(figsize=(10, 10))

# 设置多个视角
views = [(0, 0), (30, 90), (30, 135), (30, 180)]  # 4个不同视角

# 获取所有三点组合
combs = list(combinations(range(len(points)), 3))  # 从所有点中选择3个点的组合

for i, (elev, azim) in enumerate(views):
    # 创建子图
    ax = fig.add_subplot(2, 2, i+1, projection='3d')  # 2行2列的子图
    ax.set_title(f'视角 {i+1}: elev={elev}, azim={azim}')
    
    # 绘制点
    ax.scatter(points[:, 0], points[:, 1], points[:, 2], color='r', s=10)

    # 绘制每一个三角形
    for comb in combs:
        # 获取当前组合的三个点
        p1, p2, p3 = points[comb[0]], points[comb[1]], points[comb[2]]
        # 使用这些点构建一个三角形
        ax.plot_trisurf([p1[0], p2[0], p3[0]], [p1[1], p2[1], p3[1]], [p1[2], p2[2], p3[2]], color='b', alpha=0.1)

    # 设置视角
    ax.view_init(elev=elev, azim=azim)

# 显示所有视角的图形
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from itertools import combinations

# 输入点的坐标 (n = 4)
points = np.array([
    [1, 0, 1],
    [0, 5, 1],
    [0, 0, 1],
    [2, 2, 0],
])

# 创建三维图形
fig = plt.figure(figsize=(10, 10))

# 设置多个视角
views = [(30, 45), (60, 90), (90, 135), (120, 180)]  # 4个不同视角

# 获取所有三点组合
combs = list(combinations(range(len(points)), 3))  # 从所有点中选择3个点的组合

# 正方形柱体底面和顶面的点 (空心柱)
bottom_points = np.array([
    [1, 0, 0],
    [0, 0, 0],
    [0, 1, 0],
    [1, 1, 0]
])

top_points = bottom_points + np.array([0, 0, 1])  # 顶面点

# 创建一个子图
for i, (elev, azim) in enumerate(views):
    ax = fig.add_subplot(2, 2, i+1, projection='3d')  # 2行2列的子图
    ax.set_title(f'视角 {i+1}: elev={elev}, azim={azim}')
    
    # 绘制点，使用点的z坐标作为颜色的依据
    z_values = points[:, 2]  # 取点的z坐标作为颜色深浅的依据
    scatter = ax.scatter(points[:, 0], points[:, 1], points[:, 2], c=z_values, cmap='viridis', s=50)

    # 绘制每一个三角形
    for comb in combs:
        p1, p2, p3 = points[comb[0]], points[comb[1]], points[comb[2]]
        ax.plot_trisurf([p1[0], p2[0], p3[0]], [p1[1], p2[1], p3[1]], [p1[2], p2[2], p3[2]], color='b', alpha=0.5)

    # 绘制空心正方形柱的边框
    # 连接底面和顶面点形成边框
    for j in range(4):
        ax.plot([bottom_points[j, 0], top_points[j, 0]], 
                [bottom_points[j, 1], top_points[j, 1]], 
                [bottom_points[j, 2], top_points[j, 2]], color='r')
    
    # 连接底面的四个点
    for j in range(4):
        ax.plot([bottom_points[j, 0], bottom_points[(j+1)%4, 0]], 
                [bottom_points[j, 1], bottom_points[(j+1)%4, 1]], 
                [bottom_points[j, 2], bottom_points[(j+1)%4, 2]], color='r')
    
    # 连接顶面的四个点
    for j in range(4):
        ax.plot([top_points[j, 0], top_points[(j+1)%4, 0]], 
                [top_points[j, 1], top_points[(j+1)%4, 1]], 
                [top_points[j, 2], top_points[(j+1)%4, 2]], color='r')

    # 设置视角
    ax.view_init(elev=elev, azim=azim)

    # 显示图例 (只在第一个视图显示一次)
    if i == 0:
        fig.colorbar(scatter, ax=ax, shrink=0.5, aspect=5)  # 为颜色添加颜色条

# 显示所有视角的图形
plt.tight_layout()
plt.show()

```
